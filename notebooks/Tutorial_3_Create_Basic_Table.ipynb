# Tutorial 3: Create a simple table of Wikipedia page content

### Retrieve contents of soup object in earlier tutorial (or rerun Tutorial 2)

```python
%store -r soup
```

## Create a set of objects to populate your table

```python
url=""
text_title=""
text_content=""
```

## Parse soup object

## URL
### A canonical link refers to the  URL that the page is linked to

```python
for link in soup.find_all("link",{"rel":["canonical"]}):
    url=link.attrs["href"]

    
```

## Title
### The title of the Wikipedia page is identified by a class of mw-body-header and a h1 tag

```python
for h1 in soup.find("header",{"class":"mw-body-header"}).find("h1"):
    text_title = h1.get_text()
```

## Content
### The content of the page is taken from a div with id bodyContent

```python
for body in soup.find_all("div", {"id":["bodyContent"]}):
    text_content = body.get_text()
```

## Save objects in a dict structure

```python
wiki_content = {"url":[url],"Title":[text_title],"Content":[text_content]}
```

## Optional - convert to a dataframe

```python
import pandas as pd
wiki_content_pdf = pd.DataFrame(wiki_content)
```
