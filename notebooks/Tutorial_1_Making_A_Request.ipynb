# Tutorial 1: Make a web request to GET a page

## Import required packages

```python
import requests
```

## Create a reference to the Random article URL of Wikipedia

```python
random_wiki_url="https://en.wikipedia.org/wiki/Special:Random"
```

## Create header records to make yourself known to the Wikipedia server page

```python
headers = {
    'User-Agent': 'Mozilla/5.0 MyWikiRandom/1.0 (https://your-website.com; your_email_id@email.service)'
}
```

## Call the specified URL

```python
resp = requests.get(url=random_wiki_url, headers=headers)
```

### Check response

```python
if resp.ok is True:
    print("Success. Page retrieved.")
else:
    print(f"Failure. Check status. {resp.status_code} \n\n {resp.content.decode()} \n\n ")
```
