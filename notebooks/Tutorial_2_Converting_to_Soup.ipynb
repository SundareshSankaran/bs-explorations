## Import Beautiful Soup

```python
from bs4 import BeautifulSoup 
```

## Convert the response object to a soup object

```python
soup = BeautifulSoup(resp.content, 'html.parser')
```

## Extract text from the soup object

```python
# print(soup.get_text())
```

### Quick illustration 1: extract from all specified tags (divs) of a specified class

```python
for div in soup.find_all("div",{"class":"mw-content-container"}):
    print(div.get_text()[:200])
```

### Quick illustration 2: Extract the link of a Wikipedia page

```python
for link in soup.find_all("link",{"rel":["canonical"]}):
    print(link)
    url = link.attrs["href"]
    print(url)
```
